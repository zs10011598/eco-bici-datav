```python
import pandas as pd
import folium
from folium import plugins
```

```python
with open('../data/stations.json', 'r') as f:
    stations_json = f.read()

df_stations = pd.read_json(stations_json, orient='records')
df_stations['zip'] = df_stations['zip'].fillna(0) 
df_stations = df_stations.astype({'zip': int})
df_stations.set_index('id', inplace=True)

df_stations.head()
```

```python
stations_map = folium.Map(location=[19.433571, -99.167809], zoom_start=12)

for longitud, latitud in zip(df_stations['lon'], df_stations['lat']):
    folium.vector_layers.CircleMarker(
            [latitud, longitud],
            radius=3,
            color='red',
            fill=True,
            fill_color='darkred',
            fill_opacity=0.6
        ).add_to(stations_map)
```
